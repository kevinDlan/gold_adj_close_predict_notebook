* Création de feature Supplementaire qui pouront nous aider a mieux prédire le prix de l'OR
  (INDICATEUR TECHNIQUE)

```python
def calculate_MACD(data, nslow=26, nfast=12):
    emaslow = data.ewm(span=nslow, min_periods=nslow, adjust=True, ignore_na=False).mean()
    emafast = data.ewm(span=nfast, min_periods=nfast, adjust=True, ignore_na=False).mean()
    dif = emafast - emaslow
    MACD = dif.ewm(span=9, min_periods=9, adjust=True, ignore_na=False).mean()
    return dif, MACD

def calculate_RSI(data, periods=14):
    # wilder's RSI
    delta = data.diff()
    up, down = delta.copy(), delta.copy()

    up[up < 0] = 0
    down[down > 0] = 0

    rUp = up.ewm(com=periods,adjust=False).mean()
    rDown = down.ewm(com=periods, adjust=False).mean().abs()

    rsi = 100 - 100 / (1 + rUp / rDown)
    return rsi

def calculate_SMA(data, peroids=15):
    SMA = data.rolling(window=peroids, min_periods=peroids, center=False).mean()
    return SMA

def calculate_BB(data, peroids=15):
    STD = data.rolling(window=peroids,min_periods=peroids, center=False).std()
    SMA = calculate_SMA(data)
    upper_band = SMA + (2 * STD)
    lower_band = SMA - (2 * STD)
    return upper_band, lower_band

def calculate_stdev(data,periods=5):
    STDEV = data.rolling(periods).std()
    return STDEV
```

```python
GLD_adj_close.rolling(window=15, min_periods=15, center=False).mean()
```

```python
SMA_GLD = calculate_SMA(data=GLD_adj_close)
upper_band, lower_band = calculate_BB(data=GLD_adj_close)
DIF, MACD = calculate_MACD(data=GLD_adj_close)
RSI = calculate_RSI(data=GLD_adj_close)
STDEV= calculate_stdev(data=GLD_adj_close)
Open_Close=datas.Open - datas.Close
High_Low=datas.High-datas.Low
```

```python
data = datas
data['SMA'] = SMA_GLD
data['Upper_band'] = upper_band
data['Lower_band'] = lower_band
data['DIF'] = DIF
data['MACD'] = MACD
data['RSI'] = RSI
data['STDEV'] = STDEV
data['Open_Close']=Open_Close
data['High_Low']=High_Low
```
